## CREATING A USER-ITEM INTERACTION MATRIX


### Why Create the User-Item Interaction Matrix?

The User-Item Interaction Matrix is a pivotal component in our movie recommendation system for the following key reasons:

- **Representation of User-Movie Interactions:** This matrix provides a structured overview of user-movie interactions. Rows represent users, columns represent movies, and the matrix cells contain user ratings. This representation is essential for understanding user preferences.

- **Handling Sparse Data:** Real-world user rating data is typically sparse, with users rating only a fraction of available movies. The matrix handles this sparsity gracefully, with missing values indicating unrated movies.

- **Personalization:** By analyzing user preferences within the matrix, we personalize recommendations. Users are recommended movies based on their unique rating history, enhancing the user experience.

- **Similarity and Prediction:** The matrix serves as the foundation for similarity calculations. User similarity helps identify like-minded users for collaborative filtering, while movie similarity assists in content-based recommendations. Moreover, it enables predictive modeling for unrated movies.

Creating this matrix is a critical step that empowers our recommendation system to offer tailored movie suggestions and enhance user satisfaction.


### Create the User-Item Interaction Matrix

This matrix will have users as rows, movies as columns, and ratings as values. We'll assume that a missing value in the matrix means the user hasn't rated that movie.

```python
# Create the user-item interaction matrix
user_item_matrix_df = ratings.pivot(index='userId', columns='movieId', values='rating')

# Display the user-item interaction matrix
print(user_item_matrix_df)
```

## Handling Sparse Data with Mean Imputation

It's evident that the user-item interaction matrix is sparse, containing numerous missing values. These missing values arise because users have not rated every movie in the dataset.

To address this sparsity issue and ensure we can effectively work with the data, we employ a technique known as mean imputation.

Mean imputation involves filling in the missing values with the mean rating for each movie. This approach serves two essential purposes:

1. **Preserving Sparsity Structure:** Mean imputation allows us to maintain the original sparsity structure of the user-item interaction matrix. Instead of replacing missing values with arbitrary estimates, we use a value that is representative of the overall user sentiment for each movie.

2. **Bias-Free Handling:** Importantly, mean imputation avoids introducing strong biases into the dataset. By using the mean rating for a movie, we provide a reasonable estimate for unrated movies without favoring specific user groups or genres. This maintains the integrity of our recommendation system.

By implementing mean imputation, we prepare the data for further analysis and recommendation modeling, ensuring that missing ratings do not hinder the effectiveness of our movie recommendations.


```python
# Fill missing values with the mean rating for each movie
user_item_matrix = user_item_matrix_df.fillna(user_item_matrix_df.mean())

# Display the filled user-item interaction matrix
print(user_item_matrix)
```

```python
user_item_matrix.isnull().sum(), user_item_matrix.describe()
```

### Normalizing the User-Item Interaction Matrix

Now that we've effectively handled the missing values in our user-item interaction matrix, it's time to take another crucial step in our movie recommendation project: normalization.

**Why Normalize the Ratings?**

To enhance the quality of our movie recommendations, it's a common practice to normalize the ratings. This involves subtracting the mean rating of each user from their individual ratings for movies. The primary objective of this normalization process is to center the ratings distribution around zero.

Here's why we do this:

1. **User-Centric Recommendations:** Normalization makes our recommendation system more user-centric. By subtracting each user's mean rating from their ratings, we're essentially adjusting for individual user biases. This ensures that recommendations are tailored to the specific preferences of each user.

2. **Comparability:** Normalization makes it easier to compare user ratings across the entire dataset. Without normalization, users who tend to give higher ratings for all movies might dominate the recommendations, while users with lower average ratings might be overlooked. Normalizing the ratings places all users on a level playing field.

3. **Improved Model Performance:** Many recommendation algorithms perform better when working with normalized data. It can lead to more accurate and relevant movie recommendations by reducing the impact of user-specific rating tendencies.

By normalizing our user-item interaction matrix, we're taking a critical step toward building a more personalized and effective movie recommendation system. This process sets the stage for the application of various recommendation algorithms, ultimately improving the overall quality of movie suggestions for our users.


```python
# Calculate the mean rating for each user
user_mean_ratings = user_item_matrix_df.mean(axis=1)

# Normalize the user-item matrix by subtracting the mean rating of each user
normalized_user_item_matrix = user_item_matrix.sub(user_mean_ratings, axis=0)

# Display the normalized user-item matrix
print(normalized_user_item_matrix)
```

The matrix has been normalized let us now proceed to:

#### Calculate User Similarity using cosine similarity

we will calculate the similarity between users based on their normalized ratings. We'll use cosine similarity for this purpose.

we will also handle **Zero Division** in this part. This are cases where users have no common-rated items

### Calculating User Similarity using Cosine Similarity

Now that we have successfully normalized our user-item interaction matrix, which resulted in a matrix with dimensions [610 rows x 9724 columns], we can proceed to the next crucial step in our movie recommendation project: calculating user similarity.

**Why Calculate User Similarity?**

User similarity is a fundamental concept in recommendation systems, as it allows us to identify users with similar tastes and preferences. By finding users who share similar movie preferences, we can provide more relevant and personalized movie recommendations. Here's why we are doing this:

1. **Personalization:** Calculating user similarity enables us to personalize recommendations. Users who have similar tastes are likely to enjoy the same movies. By identifying these similarities, we can suggest movies that similar users have liked but the target user hasn't seen yet.

2. **Enhanced Recommendation Quality:** Cosine similarity is a popular method for measuring user similarity because it takes into account the direction and magnitude of user preferences. Users with a higher cosine similarity are more similar in their rating patterns, and this information helps us make more accurate movie recommendations.

3. **Handling Zero Division:** In this step, we also address cases where users have no common-rated items. These situations can arise when two users have not rated any of the same movies. To avoid zero division errors and provide meaningful recommendations even in these cases, we'll implement techniques to handle such scenarios gracefully.

By calculating user similarity using cosine similarity and addressing zero division cases, we are laying the foundation for a recommendation system that understands user preferences, promotes personalization, and enhances the overall quality of movie suggestions. This step is crucial in making our movie recommendation system more effective and user-centric.


```python
# Calculate the similarity matrix between users
user_similarity = cosine_similarity(normalized_user_item_matrix)

# Create a DataFrame for the user similarity matrix and replace NaN values (resulting from zero division) with 0
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index).fillna(0)
# Now you have the user similarity matrix with zero division handled
user_similarity_df
```

## 5. Building User-Based Movie Recommendations

Now that we have successfully calculated the user similarity using cosine similarity, resulting in a user-user similarity matrix of dimensions [610 rows x 610 columns], we can proceed to the exciting part of our movie recommendation project: building personalized movie recommendations for users. 

**Why Build User-Based Recommendations?**

Building user-based movie recommendations is a crucial step in our project because it leverages the insights gained from user similarity. Here's why we are doing this:

1. **Personalized Movie Suggestions:** By considering how similar one user is to others in terms of their movie preferences, we can provide personalized movie recommendations. Users tend to trust and engage more with systems that offer content tailored to their tastes, enhancing their overall experience.

2. **Expanding Movie Discovery:** Recommending movies based on user similarity can introduce users to films they might not have discovered otherwise. It broadens their horizons by suggesting movies that like-minded users have enjoyed.

**Understanding the Recommendation Function**

We've defined a Python function called `get_user_recommendations(user_id, N=10)` to retrieve the top N movie recommendations for a given user. Here's how the function works:

- **Input Parameters:**
  - `user_id`: The ID of the user for whom we want to generate recommendations.
  - `N` (optional): The number of movie recommendations to retrieve (default is 10).

- **Function Steps:**
  1. We start by fetching the normalized ratings of the specified user (`user_id`) from the normalized user-item interaction matrix.

  2. Next, we calculate the weighted average of ratings based on user similarity. This step involves multiplying the user's similarity scores with all users' normalized ratings. This process helps in identifying movies that similar users have rated highly.

  3. To ensure that we recommend movies the user hasn't already seen, we zero out the weighted ratings for movies the user has already rated.

  4. Finally, we retrieve the top N movie recommendations based on the highest weighted ratings and return their titles.



```python
# Define a function to get top N movie recommendations for a user
def get_user_recommendations(user_id, N=10):
    # Get the user's normalized ratings
    user_ratings = normalized_user_item_matrix.loc[user_id]
    
    # Calculate the weighted average of ratings by similarity
    user_similarities = user_similarity_df[user_id]
    weighted_ratings = user_similarity_df[user_id].dot(normalized_user_item_matrix)
    
    # Exclude movies the user has already rated
    rated_movies = normalized_user_item_matrix.loc[user_id].notna()
    weighted_ratings[rated_movies] = 0
    
    # Get the top N movie recommendations
    top_movie_ids = weighted_ratings.sort_values(ascending=False).index[:N]
    top_movie_titles = movies[movies['movieId'].isin(top_movie_ids)]['title']
    
    return top_movie_titles

# Example: Get top 10 movie recommendations for user with userId 1
user_id = 1
top_recommendations = get_user_recommendations(user_id, N=10)
print(f"Top 10 Movie Recommendations for User {user_id}:\n")
for i, movie_title in enumerate(top_recommendations, 1):
    print(f"{i}. {movie_title}")
```

## Evaluating the Movie Recommendation System

it's crucial to evaluate the performance of our recommendation system to ensure that it provides high-quality and accurate movie suggestions to users. We conduct this evaluation for several reasons:

1. **Measuring Recommendation Quality:** The generated movie recommendations are valuable, but we need to assess their relevance and quality. Does the system suggest movies that users are likely to enjoy? Evaluation helps answer this question.

2. **Quantifying Recommendation Accuracy:** We employ evaluation metrics such as Root Mean Square Error (RMSE) and Mean Absolute Error (MAE) to quantify the accuracy of our recommendations. Lower RMSE and MAE values indicate better accuracy.

3. **User-Centric Evaluation:** Our evaluation is user-centric, allowing us to assess how well the system performs for each individual user. This approach ensures personalized recommendations that align with user preferences.

4. **Benchmarking and Improvement:** Regular evaluation serves as a benchmark for our system's performance. Over time, as we make enhancements or algorithmic changes, we can compare new results to the benchmark to gauge improvements.

### Understanding the Evaluation Process

Here's a breakdown of the steps involved in our evaluation process:

1. **Data Splitting:** We split the user-item matrix into training and testing sets using `train_test_split`. This division enables us to assess the system's performance on unseen data, mimicking real-world scenarios.

2. **Evaluation Function:** We have defined a Python function called `evaluate_recommendations`. This function takes the training matrix, testing matrix, user similarity data, and the recommendation function as input. It performs the following tasks:
   - For each user in the test set, it retrieves the actual ratings for movies.
   - It obtains the top N recommended movies for that user using the recommendation function.
   - It filters the actual ratings for the recommended movies.
   - RMSE and MAE are calculated to measure recommendation accuracy.
   - The RMSE and MAE scores are averaged to derive system-level evaluation metrics.

3. **Example Usage:** We provide an example of how to use this evaluation function to assess the system's performance. In this example, we evaluate the system for 10-item recommendations.

4. **Output and Interpretation:** The code outputs the average RMSE and MAE scores, representing the overall accuracy of movie recommendations. Lower scores indicate higher accuracy, indicating that the recommendations align more closely with user preferences.

This evaluation process is instrumental in assessing the effectiveness of our movie recommendation system, ensuring user satisfaction, and guiding future enhancements in algorithmic performance.


```python
from sklearn.model_selection import train_test_split

# Split the user-item matrix into training and testing sets
train_matrix, test_matrix = train_test_split(normalized_user_item_matrix, test_size=0.2, random_state=42)

# Define a function to evaluate the recommendation system
def evaluate_recommendations(train_matrix, test_matrix, user_similarity_df, get_user_recommendations, N=10):
    # Initialize lists to store evaluation results
    rmse_scores = []
    mae_scores = []
    
    # Iterate through each user in the test set
    for user_id in test_matrix.index:
        # Get the actual ratings for the user from the test set
        actual_ratings = test_matrix.loc[user_id].values
        
        # Get the top N recommended movies for the user
        top_recommendations = get_user_recommendations(user_id, N=N)
        
        # Filter the actual ratings for the recommended movies
        actual_ratings = actual_ratings[top_recommendations.index]
        
        # Calculate RMSE and MAE for this user
        rmse = np.sqrt(mean_squared_error(actual_ratings, np.zeros(N)))
        mae = mean_absolute_error(actual_ratings, np.zeros(N))
        
        # Append the scores to the lists
        rmse_scores.append(rmse)
        mae_scores.append(mae)
    
    # Calculate the average RMSE and MAE scores
    avg_rmse = np.mean(rmse_scores)
    avg_mae = np.mean(mae_scores)
    
    return avg_rmse, avg_mae

# Example: Evaluate the recommendation system
N = 10  # Number of recommendations
avg_rmse, avg_mae = evaluate_recommendations(train_matrix, test_matrix, user_similarity_df, get_user_recommendations, N=N)

print(f"Average RMSE for {N}-item recommendations: {avg_rmse:.2f}")
print(f"Average MAE for {N}-item recommendations: {avg_mae:.2f}")
```

## 9. Content-Based Movie Recommendation System

### Combining and Preprocessing Movie Data

In this section, we prepared the movie data by merging relevant information from the 'movies' and 'movie_features' datasets. We also cleaned and preprocessed the text data, specifically the movie titles, genres, and tags. The goal is to make this data suitable for feature extraction.

```python
# Combine relevant information from 'movies' and 'movie_features'
movie_data = movies.merge(movie_features.groupby('movieId')['tag'].apply('|'.join).reset_index(), on='movieId', how='left')

# Clean and preprocess the text data (movie titles and genres)
movie_data['genres'] = movie_data['genres'].str.replace('|', ' ')
movie_data['title'] = movie_data['title'].str.replace(r'\(.*\)', '').str.strip()

# Display the preprocessed movie data
movie_data.head()
```

### Feature Extraction using TF-IDF

To enable content-based recommendations, we needed to convert the text data (genres, tags) into numerical representations. We utilized TF-IDF (Term Frequency-Inverse Document Frequency) to extract features from the text data:

1. **Handling Missing Values**: We filled missing values in the 'tag' column with empty strings to ensure compatibility with the TF-IDF vectorizer.

2. **TF-IDF Vectorization**: We created a TF-IDF vectorizer and fitted it to the combined text data of movie genres and tags. This transformation generated a TF-IDF matrix, providing numerical representations of movies based on their content.

```python
# Fill missing values in the 'tag' column with an empty string
movie_data['tag'].fillna('', inplace=True)

# Create a TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform the text data to obtain TF-IDF features
tfidf_matrix = tfidf_vectorizer.fit_transform(movie_data['genres'] + ' ' + movie_data['tag'])

# Display the TF-IDF matrix
tfidf_matrix.shape
```

### Calculating Movie Similarity

With the TF-IDF matrix at our disposal, we calculated the cosine similarity between movies. This similarity score reflects how alike or dissimilar two movies are based on their content. The resulting cosine similarity matrix allows us to find movies that are content-wise similar to each other.

```python
# Calculate cosine similarity between movies
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Display the shape of the cosine similarity matrix
cosine_sim.shape
```

### Recommender Function

To offer movie recommendations, we created a Python function that takes a movie title as input and returns a list of recommended movies based on content similarity. Here's a breakdown of the function's workflow:

1. **Input Movie Index**: The function identifies the index of the input movie in the dataset.

2. **Similarity Scores**: It retrieves the similarity scores between the input movie and all other movies.

3. **Sorting**: The function sorts movies by their similarity scores in descending order, with the most similar movies at the top.

4. **Excluding Input Movie**: To prevent recommending the input movie itself, it removes the first entry (which corresponds to the input movie).

5. **Top N Recommendations**: Finally, the function selects the top N recommended movies based on their similarity scores.

### Example Usage

To demonstrate the content-based recommendation system, we used the movie "Toy Story (1995)" as an example. The system returned the top 10 recommended movies with their titles.

This system leverages the content of movies (genres, tags) to suggest similar movies to users based on their preferences.

```python
def get_content_based_recommendations(movie_title, cosine_sim, top_n=10):
    # Find the index of the input movie
    movie_index = movie_data.index[movie_data['title'] == movie_title].tolist()[0]

    # Get the similarity scores for the input movie
    sim_scores = list(enumerate(cosine_sim[movie_index]))

    # Sort movies by similarity scores in descending order
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Exclude the input movie itself from recommendations
    sim_scores = sim_scores[1:]

    # Get the movie indices of the top N recommendations
    movie_indices = [x[0] for x in sim_scores[:top_n]]

    # Get the titles of the top N recommended movies
    recommended_movies = movie_data['title'].iloc[movie_indices]

    return recommended_movies

# Example usage:
input_movie_title = "Toy Story (1995)"
recommendations = get_content_based_recommendations(input_movie_title, cosine_sim)
print(recommendations)
```
