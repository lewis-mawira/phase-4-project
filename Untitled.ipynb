### Create the User-Item Interaction Matrix

This matrix will have users as rows, movies as columns, and ratings as values. We'll assume that a missing value in the matrix means the user hasn't rated that movie.

```python
# Create the user-item interaction matrix
user_item_matrix = ratings.pivot(index='userId', columns='movieId', values='rating')
```

```python
user_item_matrix
```

As you can see, the matrix is sparse, with many missing values indicating that users have not rated certain movies.

#### lets proceed no to normalizing the user matrix now.

To improve the quality of recommendations, it's common to normalize the ratings by subtracting the mean rating of each user from their ratings. This centers the ratings around zero.

```python
# Normalize the user-item interaction matrix by subtracting user means
user_item_matrix_normalized = user_item_matrix.sub(user_item_matrix.mean(axis=1), axis=0).fillna(0)
user_item_matrix_normalized
```

The matrix has been normalized let us now proceed to:

#### Calculate User Similarity

we will calculate the similarity between users based on their normalized ratings. We'll use cosine similarity for this purpose.

we will also handle **Zero Division** in this part. This are cases where users have no common-rated items

```python
# Calculate cosine similarity between users
user_similarity = cosine_similarity(user_item_matrix_normalized)

# Replace NaN values (resulting from zero division) with 0
user_similarity = pd.DataFrame(user_similarity, index=user_item_matrix_normalized.index, columns=user_item_matrix_normalized.index).fillna(0)

# Now you have the user similarity matrix with zero division handled
user_similarity
```

Now, we have a user-user similarity matrix that represents how similar each user is to every other user based on their ratings with zero division handled.


```python
# Define a function to get personalized movie recommendations for a user
def get_movie_recommendations(user_id, num_recommendations=10):
    # Get the user's similarity scores with all other users
    user_similarity_scores = user_similarity[user_id]
    
    # Create a list to store movie recommendations
    recommendations = []
    
    # Iterate through each movie
    for movie_id in user_item_matrix.columns:
        # Skip movies the user has already rated
        if pd.notna(user_item_matrix.loc[user_id, movie_id]):
            continue
        
        # Calculate the predicted rating for the movie
        predicted_rating = np.sum(user_similarity_scores * user_item_matrix_normalized.loc[:, movie_id]) / np.sum(np.abs(user_similarity_scores))
        
        # Append the movie details and predicted rating to the recommendations list
        movie_details = movies[movies['movieId'] == movie_id]
        title = movie_details['title'].values[0]
        recommendations.append({'movieId': movie_id, 'title': title, 'predicted_rating': predicted_rating})
    
    # Sort the recommendations by predicted rating in descending order
    recommendations.sort(key=lambda x: x['predicted_rating'], reverse=True)
    
    # Get the top 'num_recommendations' recommendations
    top_recommendations = recommendations[:num_recommendations]
    
    # Convert the list of recommendations to a DataFrame
    top_recommendations_df = pd.DataFrame(top_recommendations)
    
    return top_recommendations_df

# Example: Get movie recommendations for user 1
user_id = 1
recommended_movies = get_movie_recommendations(user_id)
print(f"Top 10 Movie Recommendations for User {user_id}:\n")
print(recommended_movies)

```
